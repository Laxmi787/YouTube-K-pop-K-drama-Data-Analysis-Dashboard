### Trend Tracking with Hashtags & Titles


1. What are the top hashtags over time?

```python
# Ensure datetime format
data['published_at'] = pd.to_datetime(data['published_at'], utc=True)
data['year_month'] = data['published_at'].dt.to_period('M').astype(str)

# Expand hashtags into rows
df_hashtags = data[['year_month', 'hashtags']].dropna()
df_hashtags = df_hashtags[df_hashtags['hashtags'].str.strip() != '']

# Split, clean, and explode
df_hashtags = df_hashtags.assign(hashtag=df_hashtags['hashtags'].str.split(',')).explode('hashtag')
df_hashtags['hashtag'] = df_hashtags['hashtag'].str.strip().str.lower().str.replace('#', '', regex=False)

# Count top hashtags overall
top_hashtags = df_hashtags['hashtag'].value_counts().head(10).index

# Filter top 10 hashtags
filtered = df_hashtags[df_hashtags['hashtag'].isin(top_hashtags)]
filtered_grouped = filtered.groupby(['year_month', 'hashtag']).size().reset_index(name='count')

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=filtered_grouped, x='year_month', y='count', hue='hashtag', marker='o')
plt.title('Top 10 Hashtag Trends Over Time', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Hashtag Count', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Hashtag', fontsize=9, title_fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

2. What are the emerging keywords in titles? (K-pop groups, drama names)

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer

# Convert datetime and remove timezone
data['published_at'] = pd.to_datetime(data['published_at']).dt.tz_localize(None)

# Prepare titles and month
df_titles = data[['published_at', 'title']].dropna()
df_titles['year_month'] = df_titles['published_at'].dt.to_period('M').astype(str)

# Extract keywords from titles
vectorizer = CountVectorizer(stop_words='english', max_features=1000)
keywords = vectorizer.fit_transform(df_titles['title'])

# Convert to DataFrame
keyword_df = pd.DataFrame(keywords.toarray(), columns=vectorizer.get_feature_names_out())
keyword_df['year_month'] = df_titles['year_month'].values

# Aggregate keyword counts by month
monthly_keywords = keyword_df.groupby('year_month').sum()

# Pick top 10 keywords from latest month
latest_month = monthly_keywords.index[-1]
top_keywords = monthly_keywords.loc[latest_month].sort_values(ascending=False).head(10).index

# Plot trends
trend_data = monthly_keywords[top_keywords]
trend_data.plot(figsize=(10, 5), marker='o')
plt.title('Trending Keywords in Video Titles Over Time')
plt.xlabel('Month')
plt.ylabel('Keyword Frequency')
plt.xticks(rotation=45)
plt.legend(title='Keyword')
plt.tight_layout()
plt.show()
```

3. What trending topics have high audience sentiment?


```python
from textblob import TextBlob
import pandas as pd
import matplotlib.pyplot as plt

# Drop rows with missing or empty top_comments
df_sentiment = data[['top_comments', 'hashtags']].dropna()
df_sentiment = df_sentiment[df_sentiment['top_comments'].str.strip() != '']

# Calculate sentiment polarity
df_sentiment['sentiment'] = df_sentiment['top_comments'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Expand comma-separated hashtags
df_sentiment = df_sentiment.assign(hashtag=df_sentiment['hashtags'].str.split(',')).explode('hashtag')
df_sentiment['hashtag'] = df_sentiment['hashtag'].str.strip().str.lower()

# Group by hashtag and calculate average sentiment
hashtag_sentiment = df_sentiment.groupby('hashtag')['sentiment'].mean().sort_values(ascending=False)

# Top 10 hashtags with most positive sentiment
top_positive = hashtag_sentiment.head(10)

# Convert to DataFrame for display
top_positive_df = top_positive.reset_index()
top_positive_df.columns = ['Hashtag', 'Average Sentiment']

# Display as table
print("Top 10 Hashtags by Average Positive Sentiment (Based on Top Comments):")
print(top_positive_df.to_string(index=False))

# Plot the results
top_positive.plot(kind='barh', figsize=(10, 6), color='green')
plt.title('Top Hashtags by Positive Sentiment (via Top Comments)')
plt.xlabel('Avg Sentiment')
plt.ylabel('Hashtag')
plt.tight_layout()
plt.show()
```

### Viewer Sentiment Analysis


1. What is the overall sentiment of the viewer comments (positive, neutral, negative)?

```python
from textblob import TextBlob

# Combine pinned and top comments
data['all_comments'] = data['pinned_comment'].fillna('') + ' ' + data['top_comments'].fillna('')

# Filter out rows with no comments
df = data[data['all_comments'].str.strip() != '']

# Compute sentiment polarity
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

df['sentiment_score'] = df['all_comments'].apply(get_sentiment)

# Classify sentiment
def label_sentiment(score):
    if score > 0.1:
        return 'Positive'
    elif score < -0.1:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment_label'] = df['sentiment_score'].apply(label_sentiment)

# Count sentiment categories
sentiment_counts = df['sentiment_label'].value_counts()
sentiment_counts
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart of sentiment categories
plt.figure(figsize=(8, 5))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette='pastel')

plt.title('Sentiment Distribution of Viewer Comments')
plt.xlabel('Sentiment')
plt.ylabel('Number of Videos')
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Helper function to get top N frequent words
def get_top_words(comment_series, sentiment_label, n=10):
    comments = comment_series[comment_series['sentiment_label'] == sentiment_label]['all_comments'].dropna()
    vectorizer = CountVectorizer(stop_words='english')
    word_matrix = vectorizer.fit_transform(comments)
    word_sum = word_matrix.sum(axis=0)

    word_freq = [(word, word_sum[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
    word_freq = sorted(word_freq, key=lambda x: x[1], reverse=True)

    return pd.DataFrame(word_freq[:n], columns=['Word', 'Frequency'])

# Get top 10 words from positive and negative comments
top_positive_words = get_top_words(df, 'Positive')
top_negative_words = get_top_words(df, 'Negative')

top_positive_words, top_negative_words
```
